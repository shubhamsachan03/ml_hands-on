Autograd ex that looks like we have been doing

```python
x=torch.randn([20,1],requires_grad=True)
y=3*x-2
```

```python
w=torch.tensor([1.],requires_grad=True) # initial value of w and b
b=torch.tensor([1.],requires_grad=True)
y_hat=w*x+b
loss=torch.sum((y_hat-y)**2)
```

```python
print(loss)
```

```python
loss.backward()
```

```python
print(w.grad,b.grad)
```

with loop:

```python
lr=0.01
w=torch.tensor([1.],requires_grad=True) # initial value of w and b
b=torch.tensor([1.],requires_grad=True)
print(w.item(),b.item())
for i in range(10):
    x=torch.randn([20,1])
    y=3*x-2
    y_hat=w*x+b
    loss=torch.sum((y_hat-y)**2)
    loss.backward()
    with torch.no_grad(): # to stop further computation 
        w-=lr*w.grad
        b-=lr*b.grad
        w.grad.zero_()  # setting w and b grad =0 every time 
        b.grad.zero_()
    print(w.item(),b.item())
    
```

```python
%%time
learning_rate=0.001
N=1000
epochs=2000

w=torch.randn([N], requires_grad=True)
b=torch.ones([1], requires_grad=True)
print(torch.mean(w).item(), b.item())

for i in range(epochs):
    x = torch.randn([N])
    y = torch.dot(3 * torch.ones([N]) ,x)-2

    y_hat=torch.dot(w,x)+b
    loss = torch.sum((y_hat - y)**2)

    loss.backward()

    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

        w.grad.zero_()
        b.grad.zero_()
    print(torch.mean(w).item(),b.item())
```
